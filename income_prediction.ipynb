```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("data_set/canada_per_capita_income.csv")
df.head()
```

```python
df.rename(columns={'per capita income (US$)': 'income'}, inplace=True)
df.head()
```

```python
plt.xlabel('year')
plt.ylabel('income (US$)')
plt.scatter(df.year, df.income, color='red', marker='+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.income)
```

```python
reg.predict(pd.DataFrame([[2025]], columns=['year']))
```

```python
plt.xlabel('year')
plt.ylabel('income (US$)')
plt.scatter(df.year, df.income, color='red', marker='+')
plt.plot(df.year, reg.predict(df[['year']]), color='blue')
plt.show()
```

```python
year = 2025
income = reg.predict(pd.DataFrame([[year]], columns=['year']))

plt.scatter(year, income, color='green', s=100)
plt.text(year, income, f"{income[0]:.0f}")
```

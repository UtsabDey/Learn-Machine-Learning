```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data_set/salaries.csv')
df.head()
```

```python
print(df.columns)
```

```python
dummies = pd.get_dummies(df[['company', 'job', 'degree']], drop_first=True)
dummies.head()
```

```python
inputs = pd.concat([df.drop(['company', 'job', 'degree', 'salary_more_then_100k'], axis='columns'), dummies], axis='columns')
target = df['salary_more_then_100k']
inputs.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2, random_state=42)

from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
print("Train Model Score:", model.score(X_train, y_train))
print("Test Model Score:", model.score(X_test, y_test))
```

```python
model.fit(inputs, target)
print("Score on whole data:", model.score(inputs, target))
```

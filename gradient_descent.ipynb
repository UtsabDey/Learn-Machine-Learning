```python
import numpy as np
import pandas as pd
import math
from sklearn.linear_model import LinearRegression
```

```python
def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 1000000 
    n = len(x)
    learning_rate = 0.0002  
    cost_previous = 0

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        
        # Calculate Cost (MSE)
        cost = (1/n) * np.sum((y - y_predicted)**2)
        
        # 2. Stopping criteria: Use math.isclose with 1e-20 threshold
        if math.isclose(cost, cost_previous, rel_tol=1e-20):
            break
        
        cost_previous = cost
        
        # Calculate Gradients
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        
        # Update weights
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
    return m_curr, b_curr, cost
```

```python
df = pd.read_csv("content/test_scores.csv")
x = np.array(df.math)
y = np.array(df.cs)
```

```python
# 4. Run the algorithm
m, b, final_cost = gradient_descent(x, y)

# 5. Run Sklearn comparison
reg = LinearRegression()
reg.fit(x.reshape(-1, 1), y)
m_sklearn = reg.coef_[0]
b_sklearn = reg.intercept_

# 6. Final Outputs
print(f"Gradient Descent Results: m = {m}, b = {b}, cost = {final_cost}")
print(f"Sklearn Results: m = {m_sklearn}, b = {b_sklearn}")
```
